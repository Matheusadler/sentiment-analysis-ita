```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
dataset = pd.read_csv('g1_clean.csv', engine='python')
```

```python
dataset.head()
```

```python
new_comment = []

# clean text
for i in range(dataset.shape[0]):
    comment = re.sub('[^a-zA-Z]',' ',dataset['Comment'][i]) # Remove non letters
    comment = comment.lower() # lower case
    comment = comment.split() # split list
    ps = PorterStemmer() # Get word root
    comment = [word for word in comment if not word in stopwords.words('portuguese')] # important words to analize
    comment =' '.join(comment)
    new_comment.append(comment)
```

```python
from polyglot.text import Text
```

```python
dataset['Comment'] = new_comment
dataset['Comment'].head()
```

```python
sentiment = []
error = []
```

```python
for row in dataset['Comment'].values:
     try:
         text = Text(row)
        
         if text.polarity > 0:
             sentiment.append(1)
         elif text.polarity < 0:
             sentiment.append(-1)
         else:
             sentiment.append(0)
     except:
         error.append(1)
         sentiment.append(0)
```

```python
dateD = []
dateM = []
dateY = []
```

```python
import dateparser
for row in dataset['Published'].values:
     day = dateparser.parse(row).day
     dateD.append(day)
     month = dateparser.parse(row).month
     dateM.append(month)
     year = dateparser.parse(row).year
     dateY.append(year)
```

```python
dataset['Year'] = dateY
dataset['Month'] = dateM
dataset['Day'] = dateD
```

```python
dataset['Sentiment'] = sentiment
```

```python
dataset.head()
```

```python
monthList = ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
dataset['Sentiment'] = dataset['Sentiment'].replace(0, 'Neutral')
dataset['Sentiment'] = dataset['Sentiment'].replace(1, 'Positive')
dataset['Sentiment'] = dataset['Sentiment'].replace(-1, 'Negative')
```

```python
dataset = dataset.drop(columns=['Unnamed: 0'])
```

```python
dataset.to_csv('comments-classified.csv')
```

```python
dataset['Sentiment'].value_counts()
```

```python
dataset.head()
```

```python
dataset['Keyword'] = [str(x).replace("'", "").replace(";", "").replace('"', "") for x in dataset['Keyword'].values]
```

```python
dataset['Author'] = [str(x).replace("'", "") for x in dataset['Author'].values]
```
